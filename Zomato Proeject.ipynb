# Step 2 - Create the data frame

```python
import pandas as pd
dataframe = pd.read_csv("Zomato data .csv")
print (dataframe)
```

# Convert the data type of column - rate

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0]
    return float(value)

dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

# Tpye of restaurant 

```python
dataframe.head()
```

```python
sns.countplot( x = dataframe['listed_in(type)'])
plt.xlabel("type of restaurant")
```

# Conclusion - Majoirity of the restaurant falls in dinning category

```python
dataframe .head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result,c="green",marker="o")
plt.xlabel("Type of restaurant",c = "red",size=20)
plt.ylabel("Votes",c="red",size=20)
```

# Conclusion - dinning restaurants as recieved maximum votes

```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'],bins=5)
plt.title("ratings distribution")
plt.show()
```

# Conclusion - the majority restaurants recieved ratings from 3.5 to 4

# Which modes recieve maxicus rating

```python
dataframe.head()
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order',y = 'rate',data = dataframe)
```

# conclusion - offline order recieved lower rating in comparison to online order

```python
dataframe.head()
```

```python
pivot_table = dataframe.pivot_table(
    index='listed_in(type)',
    columns='online_order',
    aggfunc='size',
    fill_value=0
)

sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')

plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

# CONCLUSION - Dinning restaurants primarily accept offline orders,whereas cafes primarily recieve online oeders.This suggests that client prefer to place orders in person at restaurants,but prefer online ordering at cafes
